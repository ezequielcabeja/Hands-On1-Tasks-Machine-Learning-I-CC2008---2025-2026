16. Obtain the cross-frequency table of bookings per year and month.

```python
# Criando a tabela cruzada (simples)
tabela_Cruzada1 = pd.crosstab(df['arrival_year'], df['arrival_month'])
tabela_Cruzada1
```

```python
#tabela cruzada (Margens)
tabela_Cruzada2 = pd.crosstab(df['arrival_year'], df['arrival_month'], margins=True, margins_name="Total")
tabela_Cruzada2
```

```python
#tabela cruzada (Normalização)
tabela_Cruzada2 = pd.crosstab(df['arrival_year'], df['arrival_month'], normalize='columns', margins=True, margins_name="Total")
tabela_Cruzada2
```

18. Obtain the histograms for all numeric variables.

```python
#Hist para vizualizacao rapida
# 1. Gera histogramas para todas as variáveis numéricas
df.hist(bins=20, figsize=(15, 10))
# 2. Ajusta o layout para que os nomes não fiquem sobrepostos
plt.tight_layout()
# 3. Exibe os gráficos
plt.show()
```

```python
#Uma visualizacao melhor com o seaborn
# Seleciona apenas as colunas numéricas
colunas_numericas = df.select_dtypes(include=['number']).columns
# Cria um gráfico para cada coluna
for col in colunas_numericas:
    plt.figure()
    sns.histplot(df[col], kde=True) # kde=True adiciona a linha de suavização
    plt.title(f'Histograma de {col}')
    plt.show()
```

```python
#Ou, um grafico com melhor visualizacao
# Numerical features
# Define a custom palette for consistent colors
palette = {'Not_Canceled': '#1f77b4', 'Canceled': '#d62728'}  # Blue for 'Not_Canceled', red for 'Canceled'

# Plot all numerical features as countplots
plt.figure(figsize=(16, len(colunas_numericas) * 3))
for i, col in enumerate(colunas_numericas, 1):
    plt.subplot(len(colunas_numericas), 1, i)
    sns.countplot(data=df, x=col, hue='booking_status', palette=palette)
    plt.title(f'Count Plot of {col} by booking_status')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.legend(title='booking_status', loc='upper right')  # Ensure legend is consistent
plt.tight_layout()
plt.show()
```
